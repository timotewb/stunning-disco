```python
import ollama
import weaviate

client: weaviate.WeaviateClient = weaviate.connect_to_local()
collection = client.collections.use(name="docs")
```

```python
# An example prompt
from ollama import EmbeddingsResponse


prompt = "What animals are llamas related to?"

# Generate an embedding for the prompt and retrieve the most relevant doc
response: EmbeddingsResponse = ollama.embeddings(
  model = "all-minilm",
  prompt = prompt,
)

results = collection.query.near_vector(near_vector = response["embedding"],
                                       limit = 1)

data = results.objects[0].properties['text']
```

```python
prompt_template = f"Using this data: {data}. Respond to this prompt: {prompt}"
```

```python
# Generate a response combining the prompt and data we retrieved in step 2
output = ollama.generate(
  model = "llama2",
  prompt = prompt_template,
)

print(output['response'])
```
